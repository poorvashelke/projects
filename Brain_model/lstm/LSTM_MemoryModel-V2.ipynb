# Set Parameters for largest number, number of samples , no of epoch

```python
from random import seed
from random import randint
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from math import sqrt
from sklearn.metrics import mean_squared_error
import random, time



# generate examples of random integers and their sum
def random_sum_pairs(n_examples, n_numbers, largest):
	X, y,list_100 = list(), list(),list()
	in_pattern = n_examples
	random.seed(time.clock())
	out_pattern = [random.randint(1,largest) for _ in range(n_examples)]
    
	X.append(in_pattern)
	y.append(out_pattern)    
	# format as NumPy arrays
	list_100 = X   
	X,y = array(X), array(y)
	# normalize
	X = X.astype('float') / float(largest * n_numbers)
	y = y.astype('float') / float(largest * n_numbers)
	return X, y,list_100
    
# invert normalization
def invert(value, n_numbers, largest):
	return round(value * float(largest * n_numbers))

# generate training data
seed(1)
n_examples = 25000
n_numbers = 1
largest = 10000000
# define LSTM configuration
n_batch = 1
n_epoch = 10000
```

# Create LSTM model

```python
# create LSTM
model = Sequential()
model.add(LSTM(10, input_shape=(n_numbers, 1)))
model.add(Dense(n_examples))
model.compile(loss='mean_squared_error', optimizer='adam')
# train LSTM
```

# Create N random numbers in array "y" with one input as X.

```python

X, y,list_100 = random_sum_pairs(n_examples, n_numbers, largest)

  
```

```python
X.shape
```

```python
y.shape
```

# Set input shape to (1,1,1)

```python
X = X.reshape(1,1, 1)
```

```python
X.shape
```

# Train LSTM model.

```python
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
```

# Predict results for input [100,250,500,1000,1500,2000,5000,10000,50000,100000]

```python

result = model.predict(X, batch_size=n_batch, verbose=0)
```

# Round  values of expected output

```python
expected =[round(float(largest * n_numbers),2) * x for x in y]
```

# Round values of predicted output

```python
import decimal
predictedRound = []
predicted = [float(largest * n_numbers * x) for x in result[0]]
for i in predicted:
    predictedRound.append(round(i))
 
```

# RMS error

```python
rmse = sqrt(mean_squared_error(expected[0], predictedRound))
rmse
```
